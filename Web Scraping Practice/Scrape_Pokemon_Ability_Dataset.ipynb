```python
RGB_data = []
main_data = []


for names in gen1_pokemon:
  # Iterates through gen 1 pokemon list
  url = 'https://bulbapedia.bulbagarden.net/wiki/'+names+'_(Pokémon)/Generation_I_learnset#By_leveling_up'

  response = requests.get(url) # make request to access website


  if response.status_code == 200:
    # Parse the HTML Content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the 4th table in the website
    table = soup.find_all('table')[3]

    # Find all the rows within the table
    rows = table.find_all('tr')

    # Getting the data within each row
    for row in rows:
      row_data = []
      for cell in row.find_all(['th', 'td']):
        row_data.append(cell.text.strip())

      row_data.append(names)

      if len(row_data) == 7:
        main_data.append(row_data)
        main_data[0] = ['Level', 'Move', 'Type', 'Power', 'Accuracy', 'PP', 'Pokemon']
        main_moves = pd.DataFrame(main_data[1:], columns=main_data[0])
      elif len(row_data) == 8:
        RGB_data.append(row_data)
        RGB_data[0] = ['Level','Y', 'Move', 'Type', 'Power', 'Accuracy', 'PP', 'Pokemon']
        RGB_moves = pd.DataFrame(RGB_data[1:], columns=RGB_data[0])


  else:
    print('Not Accessible')
```

### Cleaning Main Final

```python
main_moves.head(10)
```

```python
main_df = main_moves.copy()
```

```python
main_df.info()
```

```python
main_df = main_df.fillna('NAN')

main_df = main_df.drop_duplicates()
```

```python
main_df = main_df[main_moves["Level"].str.contains("Level") == False]
```

```python
# Cleaning Level
```

```python
main_df
```

```python
main_df['Temp1'] = main_df['Level'].str.extract('\d\d(\d{2})')
main_df['Temp2'] = main_df['Level'].str.extract('^0[0-9](.)')

main_df['Level'] = main_df['Temp1'].combine_first(main_df['Temp2'])
main_df
```

```python
main_df.drop(['Temp1', 'Temp2'], axis = 1, inplace = True)
main_df
```

```python
# Cleaning Power Column
main_df['Power'] = main_df['Power'].str.extract('(..$)')
main_df['Power'] = main_df['Power'].str.replace('0—', '—')
main_df.sample(20)
```

```python
# Cleaning Accuracy Column
main_df['Temp1'] = main_df['Accuracy'].str.extract('}}(\d+%)')
main_df['Temp2'] = main_df['Accuracy'].str.extract('}}(—+%)')

main_df['Accuracy'] = main_df['Temp1'].combine_first(main_df['Temp2'])
main_df.drop(['Temp1', 'Temp2'], axis = 1, inplace = True)

main_df
```

```python
main_df['Accuracy'] = main_df['Accuracy'].str.replace('—%', '—')
main_df
```

Cleaning RGB Dataset

```python
RGB_df = RGB_moves.copy()
RGB_df
```

```python
RGB_df = RGB_df.fillna('NAN')
RGB_df = RGB_df.drop_duplicates()
```

```python
RGB_df = RGB_df[RGB_df["Level"].str.contains("RGB") == False].drop('Y', axis = 1)
```

```python
# Cleaning Accuracy Column
RGB_df['Temp1'] = RGB_df['Accuracy'].str.extract('(\d+%)')
RGB_df['Temp2'] = RGB_df['Accuracy'].str.extract('(—+%)')

RGB_df['Accuracy'] = RGB_df['Temp1'].combine_first(RGB_df['Temp2'])
RGB_df.drop(['Temp1', 'Temp2'], axis = 1, inplace = True)

RGB_df
```

```python
RGB_df['Accuracy'] = RGB_df['Accuracy'].str.replace('—%', '—')
RGB_df
```

## Finalising the Dataset

```python
Pokemon_abilities_df = pd.concat([main_df, RGB_df])
Pokemon_abilities_df
```

```python
# Create a dictionary with numbered keys
pokemon_dict = {i + 1: name for i, name in enumerate(gen1_pokemon)}

# Display the dictionary
print(pokemon_dict)
```

```python
def poke_id(chosen_item):
  '''
  returns item value for regular items
  '''
  pokemon_dict = {i + 1: name for i, name in enumerate(gen1_pokemon)}
  pokemon_dict = {v: k for k, v in pokemon_dict.items()}

  return pokemon_dict.get(chosen_item, "No info available")
```

```python
Pokemon_abilities_df['ID'] = Pokemon_abilities_df['Pokemon'].apply(poke_id)
```

```python
Pokemon_abilities_df = Pokemon_abilities_df[['ID','Pokemon', 'Level', 'Move', 'Type', 'Power', 'Accuracy', 'PP']]
Pokemon_abilities_df.sort_values(by ='ID', ascending = True)
```
